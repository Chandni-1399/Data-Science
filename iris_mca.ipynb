```python
X = chandni.drop(['species'],axis=1)
print(X)
y=chandni['species']
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=12)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_chandni= LogisticRegression()
logistic_chandni.fit(X_train,y_train)
```

```python
logistic_predict= logistic_chandni.predict(X_test)
print(logistic_predict)
```

```python
from sklearn.metrics import accuracy_score
chandni_accuracy=accuracy_score(y_test,logistic_predict)
print(chandni_accuracy)
```

```python
import sklearn.metrics as metrics
chandni_cm=metrics.confusion_matrix(y_test,logistic_predict)
print(chandni_cm)
```
